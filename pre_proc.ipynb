```python
import torch
import torchvision.transforms as transforms
from dataset import ReadImages
from PIL import Image
from os import path
```

```python
dataset = '/home/mrim/data/collection/GUIMUTEIC/FOURVIERE_CLEAN2/TRAIN_I'
dataset_test = '/home/mrim/data/collection/GUIMUTEIC/FOURVIERE_CLEAN2/TEST_I'
collection = 'fou'
out_path = './data/pre_proc/fourviere_clean2_224sq'
out_path_test = './data/pre_proc/fourviere_clean2_224sq/test'

def match(x):
    if collection == 'cli':
        return x.split('/')[-1].split('-')[0]
    else:
        s = x.split('/')[-1].split('_')
        return s[0] + s[1]
# training and test sets (scaled to 300 on the small side)
dataSetFull = ReadImages.readImageswithPattern(dataset, match)
testSetFull = ReadImages.readImageswithPattern(dataset_test, match)
```

```python
def filter_wall(im, lab):
    return 'wall' not in lab

def resize(dataset, out_path, newsize1, newsize2, f=lambda im, lab: True):
    for im, lab in dataset:
        if not f(im, lab):
            continue
        img = Image.open(im)
        img = img.resize((newsize1, newsize2), Image.BILINEAR)
        out_p = path.join(out_path, im.split('/')[-1])
        print('/'.join(im.split('/')[-3:]), '->', '/'.join(out_p.split('/')[-3:]))
        img.save(out_p)
```

```python
resize(dataSetFull, out_path, 224, 224, filter_wall)
resize(testSetFull, out_path_test, 224, 224, filter_wall)
```
